## Convert each dataframe in a column of a specific disease

```python
def df_to_colum(df_disease, disiese, year):
  # From week 1 to week 52
  for i in range(1,53):
    # If is the 1st week we need to create the new data frame
    if (i == 1):
      # Take from df_disiese rows with week 01
      cases_week_df = df_disease[df_disease.SEMANA == i][['Municipality code', 'State code', 'cod_mun', 'conteo_casos']]
      # Rename cases column with the year and week information
      cases_week_df = cases_week_df.rename(columns={'conteo_casos': 'cases_'+ disiese + '_' + str(year) + '_week_0' + str(i)})
    elif (i < 10):
      # Take from df_disiese rows with week 02 .. 09 in an auxiliar df
      cases_week_df_aux = df_disease[df_disease.SEMANA == i][['Municipality code', 'State code', 'cod_mun', 'conteo_casos']]
      # Rename cases column with the year and week information
      cases_week_df_aux = cases_week_df_aux.rename(columns={'conteo_casos': 'cases_'+ disiese + '_' + str(year) + '_week_0' + str(i)})
      # Merge the new auxiliar df with the df with past weeks outer, so we doesn't loose new municipalities
      cases_week_df = cases_week_df.merge(cases_week_df_aux, on=['Municipality code', 'State code', 'cod_mun'], how='outer')
    else:
      # Take from df_disiese rows with week 10 ... 52 in an auxiliar df
      cases_week_df_aux = df_disease[df_disease.SEMANA == i][['Municipality code', 'State code', 'cod_mun', 'conteo_casos']]
      # Rename cases column with the year and week information
      cases_week_df_aux = cases_week_df_aux.rename(columns={'conteo_casos': 'cases_'+ disiese + '_' + str(year) + '_week_' + str(i)})
      # Merge the new auxiliar df with the df with past weeks outer, so we doesn't loose new municipalities
      #cases2017_week_df = pd.concat([cases2017_week_df, cases2017_week_df_aux], join="outer")
      cases_week_df = cases_week_df.merge(cases_week_df_aux, on=['Municipality code', 'State code', 'cod_mun'], how='outer')
  disease_cases_week_df = cases_week_df.fillna(0)
  return disease_cases_week_df
```

```python
# Merge dataframes using Municipality code and 'State code'
def merge_two_df(df1, df2, how = 'outer'):
  df3 = df1.merge(df2, on=['Municipality code', 'State code', 'cod_mun'], how=how)
  df3 = df3.fillna(0)
  return df3
```

# Desnutrición

```python
def get_Desnutricion_df(sivigila_df):
  # Get year
  year = int(pd.unique(sivigila_df.ANO))
  if (year >= 2016):

    # Get dataframe of 'MORTALIDAD POR DESNUTRICIÓN'
    df_disease = sivigila_df[sivigila_df.Nombre == 'MORTALIDAD POR DESNUTRICIÓN']
    # Extract all cases as a new column
    disease_cases_week_df_1 = df_to_colum(df_disease, 'MortalidadDesnutricion', year)
    # Get dataframe of 'DESNUTRICIÓN AGUDA EN MENORES DE 5 AÑOS'
    df_disease = sivigila_df[sivigila_df.Nombre == 'DESNUTRICIÓN AGUDA EN MENORES DE 5 AÑOS']
    # Extract all cases as a new column
    disease_cases_week_df_2 = df_to_colum(df_disease, 'DesnutricionMenores5', year)
    
    # Merge 'MORTALIDAD POR DESNUTRICIÓN' and 'DESNUTRICIÓN AGUDA EN MENORES DE 5 AÑOS'
    cases_week_df = merge_two_df(disease_cases_week_df_1, disease_cases_week_df_2)

  elif (year >= 2013 and year < 2016):

    # Get dataframe of 'MORTALIDAD POR DESNUTRICIÓN'
    df_disease = sivigila_df[sivigila_df.Nombre == 'MORTALIDAD POR DESNUTRICIÓN']
    # Extract all cases as a new column and Return this values
    cases_week_df = df_to_colum(df_disease, 'MortalidadDesnutricion', year)

  return cases_week_df
```

```python
def get_two_Desnutricion_dfs(sivigila_df):
  # Get year
  year = int(pd.unique(sivigila_df.ANO))
  if (year >= 2016):
    # Get dataframe of 'MORTALIDAD POR DESNUTRICIÓN'
    df_disease = sivigila_df[sivigila_df.Nombre == 'MORTALIDAD POR DESNUTRICIÓN']
    # Extract all cases as a new column
    disease_cases_week_df_1 = df_to_colum(df_disease, 'MortalidadDesnutricion', year)
    # Get dataframe of 'DESNUTRICIÓN AGUDA EN MENORES DE 5 AÑOS'
    df_disease = sivigila_df[sivigila_df.Nombre == 'DESNUTRICIÓN AGUDA EN MENORES DE 5 AÑOS']
    # Extract all cases as a new column
    disease_cases_week_df_2 = df_to_colum(df_disease, 'DesnutricionMenores5', year)
    
    # Return two dfs
    return disease_cases_week_df_1, disease_cases_week_df_2
  
  elif (year >= 2013 and year < 2016):
    # Get dataframe of 'MORTALIDAD POR DESNUTRICIÓN'
    df_disease = sivigila_df[sivigila_df.Nombre == 'MORTALIDAD POR DESNUTRICIÓN']
    # Extract all cases as a new column
    disease_cases_week_df_1 = df_to_colum(df_disease, 'MortalidadDesnutricion', year)
    
    # Return two dfs
    return disease_cases_week_df_1
  else:
    pass


```

```python
def merge_desnuticion_dfs(desnutricion_dfs):
  for i in range(len(desnutricion_dfs)-1):
    if (i == 0):
      merged_df = merge_two_df(desnutricion_dfs[i], desnutricion_dfs[i+1])  
    else:
      merged_df = merge_two_df(merged_df, desnutricion_dfs[i+1]) 
  return merged_df
```

```python
municipalities = pd.read_csv('Data/DIVIPOLA_Municipios.csv', encoding='latin1', sep=';')[['COD_MPIO', 'NOM_MPIO']]
municipalities.rename(columns={'COD_MPIO': 'Municipality code', 'NOM_MPIO': 'Municipality'}, inplace=True)
municipalities
```

```python
import epiweeks
def get_date(name):
    """Get epi week from string"""
    epiweek = name.split('-')
    date = epiweeks.Week(int(epiweek[0]), int(epiweek[1]))
    #print(date.enddate())
    return date.enddate()
```

```python
def rename_columns(name):
    name = name[(name.index('_2') + 1): name.index('_2') + 5] + name[(name.index('_week') + 1): ]
    new_name = ''.join(i for i in name if i.isdigit() or i == '_')
    new_name = new_name.replace("_", "-")
    new_name = get_date(new_name)
    
    return new_name
```

```python
desnutricion_menores_cauca = desnutricion_menores5[(desnutricion_menores5['State code'] == 19) & (desnutricion_menores5['cod_mun'] != 0)].drop(columns=['State code', 'cod_mun'])
desnutricion_menores_cauca = municipalities.merge(desnutricion_menores_cauca, on=['Municipality code'], how='inner')
columns = desnutricion_menores_cauca.iloc[:,2:].columns.to_series().apply(rename_columns)
desnutricion_menores_cauca.columns.values[2:] = columns.values

desnutricion_menores_cauca
```

## Time Series

```python
time_series_desnutricion_menores_cauca = desnutricion_menores_cauca.T
time_series_desnutricion_menores_cauca.columns = time_series_desnutricion_menores_cauca.iloc[1]
time_series_desnutricion_menores_cauca = time_series_desnutricion_menores_cauca.iloc[2:,:]
time_series_desnutricion_menores_cauca['Cauca'] = time_series_desnutricion_menores_cauca.sum(axis=1)
time_series_desnutricion_menores_cauca
```

```python
import matplotlib.pyplot as plt
from matplotlib.dates import date2num

import seaborn as sns
plt.figure(figsize=(15, 5))
ax = sns.lineplot(x=time_series_desnutricion_menores_cauca.index, y='Cauca', data=time_series_desnutricion_menores_cauca)
ax = sns.regplot(x=time_series_desnutricion_menores_cauca.index.to_series().apply(date2num), y="Cauca", data=time_series_desnutricion_menores_cauca, x_estimator=np.mean,color="g",scatter=False)
plt.xticks(rotation=20)
ax.set(xlabel='Fecha', ylabel='Casos en el Cauca')
plt.title('Casos de desnutrición aguda en menores de 5 años en Cauca')
plt.savefig("TimeSeries-DesnutriciónMenores5-Cauca.png")
plt.show()
```
