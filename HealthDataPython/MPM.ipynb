```python
# Import necessary modules
import geopandas as gpd
```

```python
path = '/content/drive/MyDrive/GEE_GIS_Dengue/MPM/shapefile/'
```

```python
# Set filepath
fp = path + "MPM_FuenteCensal_2018.shp"
```

```python
# Read file using gpd.read_file()
data = gpd.read_file(fp)
data.head(3)
```

```python
data['Municipality code'] = data.DPTO_CCDGO.astype(str) + data.MPIO_CCDGO.astype(str)
data['Municipality code'] = data['Municipality code'].astype(int)
data = data.drop(['geometry','MPIO_CCDGO'], axis=1)
data
```

```python
data.to_csv('mpm_data_dane.csv')
```
