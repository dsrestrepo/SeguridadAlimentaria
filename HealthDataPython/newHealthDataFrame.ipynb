## undernourishment mortality

```
df_sivigilia_2017 = pd.read_csv("/content/drive/MyDrive/Nutrición/HealthDataPython/Sivigilia/SIVIGILA_2017.csv")
df_sivigilia_2017.head()
```

#Data Cleaning Sivigila

```
df_sivigilia_2017.Nombre = df_sivigilia_2017.Nombre.str.strip()
df_sivigilia_2017.nom_mun = df_sivigilia_2017.nom_mun.str.replace('(cd)','')
df_sivigilia_2017.nom_mun = df_sivigilia_2017.nom_mun.str.replace('_',' ')

df_sivigilia_2017.nom_mun = df_sivigilia_2017.nom_mun.str.replace('cd','')
df_sivigilia_2017.nom_mun = df_sivigilia_2017.nom_mun.str.replace('municipio desconocido','') ##??

df_sivigilia_2017.nom_mun = df_sivigilia_2017.nom_mun.str.replace('[^\w\s]','')


df_sivigilia_2017.nom_mun = df_sivigilia_2017.nom_mun.str.lower()
df_sivigilia_2017.nom_mun = df_sivigilia_2017.nom_mun.str.strip()
df_sivigilia_2017.head()
```

```
pd.unique(df_sivigilia_2017.Nombre)
```

```
df_mortalidad = df_sivigilia_2017[df_sivigilia_2017.Nombre == 'MORTALIDAD POR DESNUTRICIÓN']
df_mortalidad.head()
```

```
#df_mortalidad[df_mortalidad.ANO != 2017] # all from 2017
#df_mortalidad.SEMANA.max() # we have 52 weeks
#df_mortalidad.groupby('SEMANA')['nom_mun'].count().sort_values(ascending=False)[0:6]
len(pd.unique(df_mortalidad.nom_mun))
```

```
for i in range(1,53):
  if (i == 1):
    cases2017_week_df = df_mortalidad[df_mortalidad.SEMANA == i][['nom_mun', 'conteo_casos']]
    cases2017_week_df = cases2017_week_df.rename(columns={'conteo_casos': 'cases_week_'+ str(i)})
  else:
    cases2017_week_df_aux = df_mortalidad[df_mortalidad.SEMANA == i][['nom_mun', 'conteo_casos']]
    cases2017_week_df_aux = cases2017_week_df_aux.rename(columns={'conteo_casos': 'cases_week_'+ str(i)})
    #cases2017_week_df = pd.concat([cases2017_week_df, cases2017_week_df_aux], join="outer")
    cases2017_week_df = cases2017_week_df.merge(cases2017_week_df_aux, on="nom_mun", how='outer')
mortality_cases2017_week_df = cases2017_week_df
mortality_cases2017_week_df
```

```
mortality_cases2017_week_df = mortality_cases2017_week_df.fillna(0)
mortality_cases2017_week_df
```

# cases undernourishment < 5 Years

```
df_under5_2017 = df_sivigilia_2017[df_sivigilia_2017.Nombre == 'DESNUTRICIÓN AGUDA EN MENORES DE 5 AÑOS']
df_under5_2017.head()
```

```
#df_under5_2017[df_under5_2017.ANO != 2017] # all from 2017
#df_under5_2017.SEMANA.max() # we have 52 weeks
#df_under5_2017.groupby('SEMANA')['nom_mun'].count().sort_values(ascending=False)[0:6]
len(pd.unique(df_under5_2017.nom_mun))
#len(pd.unique(df_under5_2017.cod_mun))
```

```
for i in range(1,53):
  if (i == 1):
    cases2017_week_df = df_under5_2017[df_under5_2017.SEMANA == i][['nom_mun', 'conteo_casos']]
    cases2017_week_df = cases2017_week_df.rename(columns={'conteo_casos': 'cases_week_'+ str(i)})
  else:
    cases2017_week_df_aux = df_under5_2017[df_under5_2017.SEMANA == i][['nom_mun', 'conteo_casos']]
    cases2017_week_df_aux = cases2017_week_df_aux.rename(columns={'conteo_casos': 'cases_week_'+ str(i)})
    #cases2017_week_df = pd.concat([cases2017_week_df, cases2017_week_df_aux], join="outer")
    cases2017_week_df = cases2017_week_df.merge(cases2017_week_df_aux, on="nom_mun", how='outer')
df_under5_cases2017_week_df = cases2017_week_df
df_under5_cases2017_week_df
```

```
df_under5_cases2017_week_df = df_under5_cases2017_week_df.drop_duplicates()
df_under5_cases2017_week_df
```

```
df_under5_cases2017_week_df = df_under5_cases2017_week_df.fillna(0)
df_under5_cases2017_week_df
```

```
df_under5_cases2017_week_df = df_under5_cases2017_week_df.rename(columns={'nom_mun': 'Municipality'})
df_under5_cases2017_week_df
```

# Dengue df

```
sociodemographic_df = pd.read_csv("/content/drive/MyDrive/Nutrición/HealthDataPython/DengueData/Copia de DANE_Dengue_Data_csv.csv")
sociodemographic_df
```

### Clean Dengue Data

```
sociodemographic_df.Municipality = sociodemographic_df.Municipality.str.replace('á', 'a')
sociodemographic_df.Municipality = sociodemographic_df.Municipality.str.replace('é', 'e')
sociodemographic_df.Municipality = sociodemographic_df.Municipality.str.replace('í', 'i')
sociodemographic_df.Municipality = sociodemographic_df.Municipality.str.replace('ó', 'o')
sociodemographic_df.Municipality = sociodemographic_df.Municipality.str.replace('ú', 'u')
sociodemographic_df.Municipality = sociodemographic_df.Municipality.str.lower()
sociodemographic_df.Municipality = sociodemographic_df.Municipality.str.capitalize()
sociodemographic_df = sociodemographic_df.drop(['Cases 2015','Cases 2016','Cases 2017','Cases 2018'], axis=1)
sociodemographic_df.head()
```

# MPM dataframe

```
mpm_df = pd.read_csv('/content/drive/MyDrive/Nutrición/HealthDataPython/MPM_dane/mpm_data_dane.csv')
mpm_df = mpm_df.drop(['Unnamed: 0', 'DPTO_CCDGO', 'MPIO_CNMBR', 'MPIO_CCNCT', 'DPTO_CNMBR'], axis=1)
mpm_df.head()
```

## Merge MPM & Dengue

```
# Merge
sociodemographic_socioeconomic_df = sociodemographic_df.merge(mpm_df, on=['Municipality code'], how='inner')
sociodemographic_socioeconomic_df.head()
```
