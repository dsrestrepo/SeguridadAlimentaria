```python
# See the shape of all images:



def get_size(folder):
    sizes = []
    # Get from main directory all sub-directories
    for folder_path in os.listdir(folder):
        # Skip
        if (folder_path == '.DS_Store'):
            continue
        # See Train and Test sub-directories
        for folder_name in os.listdir(os.path.join(folder, folder_path)):
            # Skip
            if (folder_name in ['.DS_Store', 'test_v2_file_mapping.csv', 'train_v2.csv']):
                continue
            print(folder_name)
            for filename in os.listdir(os.path.join(folder, folder_path, folder_name)):
                
                # take image
                img = Image.open(os.path.join(folder, folder_path, folder_name, filename))
                # Get image with 
                #print(img.size)
                # Stores data like: (width, height)
                sizes.append(img.size)
            
    print(f'the max width is: {max(sizes[0])}, and the min width is: {min(sizes[0])}')
    print(f'the max height is: {max(sizes[1])}, and the min height is: {min(sizes[1])}')
    print(f'the mean width is: {np.mean(sizes[0])}, and the mean height is: {np.mean(sizes[1])}')

# Call the function
get_size(main_path)
```

# Create Dataset
* This is a multi labels task, so we need binary encode the tags
* We don't have labels for testing images so we need to split data of train in train and test
* We have many information so we need a data loader to wrap-up the data and avoid overload the GPU or RAM

## Binary Encode

```python
tags_encode = train_map.tags.str.get_dummies(sep=' ').columns
tags_encode
```

```python
# Select Tags
list_tags = ['agriculture', 'bare_ground', 'cultivation', 'habitation', 'primary', 'road', 'water']
```

## Filter images with tags 
* We have many tags, but we just want images with tags in list_tags so we filter

```python
def filter_item(tag_string):
    res = any(tag in tag_string for tag in list_tags)
    return res    
```

```python
train_map = train_map[train_map.tags.apply(lambda tag_string: filter_item(tag_string))]
train_map
```

## Binary Encode and desired columns 

```python
# Binary Encode
encode = train_map.tags.str.get_dummies(sep=' ')
train_map = pd.concat([train_map, encode], axis=1)
train_map.drop(columns=["tags"], inplace=True)
list_tags.insert(0,'image_name')
train_map = train_map[list_tags]
list_tags.pop(0)
train_map
```

### Add ".jpg" to image name

```python
train_map.image_name = train_map.image_name.apply(lambda name: name + '.jpg')
train_map
```

```python
from torch.utils.data import DataLoader, Dataset
import torch
import os

# Data Generator class
class ImageDataset(Dataset):
    'Characterizes a dataset for PyTorch'
    def __init__(self, path_dir, csv_file, image_dir, transform):
        'Initialization'
        self.path_dir = path_dir 
        self.csv_file = csv_file
        self.image_dir = os.path.join(path_dir, image_dir)
        self.transform = transform

    def __len__(self):
        'Denotes the total number of samples'
        return len(self.csv_file)

    
    def __getitem__(self, index):
        
        'Generates one sample of data'
        # Select sample
        # take image name in csv row index, and column 'image_name'
        image_name = self.csv_file.iloc[index]['image_name']
        # Get image path
        image = os.path.join(self.image_dir, image_name)
        image = Image.open(image).convert('RGB')
        
        # Transform to pytorch
        if self.transform:
            image = self.transform(image)
        
        # take image name in csv row index, and column 'image_name'
        labels = self.csv_file.iloc[index][1:].values.astype(int)
        # to pytoch
        labels = torch.tensor(labels)

        return image, labels

```

```python
train_dataset = ImageDataset(train_path, train, 'train-jpg', transform_train)
train_datagen = DataLoader(train_dataset, batch_size=32, shuffle=True)

test_dataset = ImageDataset(train_path, test, 'train-jpg', transform_train)
test_datagen = DataLoader(train_dataset, batch_size=32, shuffle=True)

validation_dataset = ImageDataset(train_path, validation, 'train-jpg', transform_train)
validation_datagen = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

```python
import torch.nn as nn
import torch.nn.functional as F
import torchvision

outs = len(list_tags)

# Resnet50
class Model(torch.nn.Module):
    def __init__(self, n_outputs=outs, pretrained=False, freeze=False):
        super().__init__()
        
        
        # Download Resnet50 pretrained if pretrained = True
        resnet = torchvision.models.resnet50(pretrained=pretrained)
        
        # Remove last layer (output)
        self.resnet = torch.nn.Sequential(*list(resnet.children())[:-1])
        
        # If Freeze = True then the model is not traiable
        # Else Model is trainable (requires_grad=True)
        if freeze:
            for param in self.resnet.parameters():
                param.requires_grad=False
        
        # add last layer
        self.flatten = nn.Flatten() 
        
        #self.fc = torch.nn.Linear(512, outs) # if ResNet 18
        self.fc = torch.nn.Linear(2048, outs) # if ResNet 50

    def forward(self, x):
        x = self.resnet(x)
        x = self.flatten(x)
        x = self.fc(x)
        #x = F.sigmoid(self.fc(x))
        return x

    def unfreeze(self):
        for param in self.resnet.parameters():
            param.requires_grad=True
```

```python
# See the model
model = Model()
model
```

```python
# If cuda
model.to(device)
```

```python
import torch.optim as optim
import numpy as np

#criterion = nn.CrossEntropyLoss()
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

for epoch in range(10):  # loop over the dataset multiple times

    running_loss = []
    steps = 0
    
    for inputs, labels in iter(train_datagen):
        steps += 1
        inputs, labels = inputs.to(device), labels.to(device)
        model.train()

        # zero the parameter gradients
        optimizer.zero_grad()
        # forward + backward + optimize
        outputs = model(inputs)
        loss = criterion(outputs, labels.float())
        loss.backward()
        optimizer.step()
        # add loss to array
        running_loss.append(loss.item())
        
        # Do validation each 20 steps in epoch % 10
        if ((steps % 20 == 0) and (epoch % 5 == 0)):
            
            model.eval()
            # Turn off gradients for validation, saves memory and computations
            with torch.no_grad():
                val_loss = []
                for val_images, val_labels in iter(validation_datagen):
                    #val_images, val_labels = val_images.to('cuda'), val_labels.to('cuda')
                    output_val = model.forward(val_images)
                    val_loss.append(criterion(output_val, val_labels.float()).item())

            print(f'Epoch {epoch} - Step: {steps}: validation loss: {np.mean(val_loss)} - train loss: {np.mean(running_loss)}')
            
    if epoch % 10 == 0:    # print every 10 epoch
        print(f'The loss are {np.mean(running_loss)}')
print('Finished Training')
```
