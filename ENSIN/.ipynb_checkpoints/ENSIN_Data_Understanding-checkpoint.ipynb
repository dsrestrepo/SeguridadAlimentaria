```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ensin_df = pd.read_csv('ENSIN output/ENSIN_final.csv')
ensin_df.rename(columns={'c_mpio': 'Municipality code'}, inplace=True)
ensin_df
```

```python
for column in ensin_df.columns:
    print(column)
```

# Municipalities

```python
municipalities = pd.read_csv('Datos/DIVIPOLA_Municipios.csv', encoding='latin1', sep=';')[['COD_MPIO', 'NOM_MPIO']]
municipalities.rename(columns={'COD_MPIO': 'Municipality code', 'NOM_MPIO': 'Municipality'}, inplace=True)
municipalities
```

# Seguridad Alimentaria

```python
columns = ['Municipality code', 'seguridad', 'menor_calidad', 'disminuir_porciones', 'respuestas_sa_1mod']

sa_df = ensin_df[columns]
sa_df = municipalities.merge(sa_df, on=['Municipality code'], how='inner')
sa_df.drop(columns=['Municipality code'], inplace=True)
sa_df = sa_df.groupby(['Municipality']).sum()#.T
#sa_df['CAUCA'] = sa_df.sum(1)
#sa_df = sa_df.T
sa_df
```

```python
sa_df['seguridad alimentaria (%)'] = sa_df['seguridad'] / sa_df['respuestas_sa_1mod'] * 100
sa_df['disminuir calidad alimentos (%)'] = sa_df['menor_calidad'] / sa_df['respuestas_sa_1mod'] * 100
sa_df['disminuir porciones alimentos (%)'] = sa_df['disminuir_porciones'] / sa_df['respuestas_sa_1mod'] * 100

sa_df.sort_values(by=['seguridad alimentaria (%)'], ascending=False)[['seguridad alimentaria (%)', 'disminuir calidad alimentos (%)', 'disminuir porciones alimentos (%)']].to_excel('Estadisticas_SA_Cauca.xlsx')
```

```python
sa_df
```

```python
top_9 = sa_df.sort_values(by=['seguridad alimentaria (%)'], ascending=False)[['seguridad alimentaria (%)', 'disminuir calidad alimentos (%)', 'disminuir porciones alimentos (%)']]
top_9
```

```python
plt.figure(figsize=(15, 5))
ax = sns.barplot(x=top_9.index , y="seguridad alimentaria (%)", data=top_9)

plt.xticks(rotation=20)
ax.set(xlabel='Municipio')
plt.title('Seguridad Alimentaria en los municipios del Cauca')
plt.savefig("SA-Cauca.png")
plt.show()
```

```python
sa_df.describe()[['seguridad alimentaria (%)', 'disminuir calidad alimentos (%)', 'disminuir porciones alimentos (%)']].T.to_excel('Estadisticas_sa_Cauca2.xlsx')
```

# Socioeconómicos

```python
columns = ['Municipality code', 'noacueducto', 'noenergia', 'Ingresos', 'respuestas_pts']

socioeconomicos_df = ensin_df[columns]
socioeconomicos_df = municipalities.merge(socioeconomicos_df, on=['Municipality code'], how='inner')
socioeconomicos_df.drop(columns=['Municipality code'], inplace=True)
socioeconomicos_df = socioeconomicos_df.groupby(['Municipality']).sum()#.T
#sa_df['CAUCA'] = sa_df.sum(1)
#sa_df = sa_df.T
socioeconomicos_df
```

```python
socioeconomicos_df['hogares con acueducto (%)'] = socioeconomicos_df['noacueducto'] / socioeconomicos_df['respuestas_pts'] * 100
socioeconomicos_df['hogares con energia (%)'] = socioeconomicos_df['noenergia'] / socioeconomicos_df['respuestas_pts'] * 100
socioeconomicos_df['ingresos insuficientes (%)'] = socioeconomicos_df['Ingresos'] / socioeconomicos_df['respuestas_pts'] * 100

socioeconomicos_df.sort_values(by=['ingresos insuficientes (%)'], ascending=False)[['ingresos insuficientes (%)', 'hogares con energia (%)', 'hogares con acueducto (%)']].to_excel('Estadisticas_socioeconómicas_Cauca.xlsx')
```

```python
top_9 = socioeconomicos_df.sort_values(by=['ingresos insuficientes (%)'], ascending=False)[['ingresos insuficientes (%)', 'hogares con energia (%)', 'hogares con acueducto (%)']]
top_9
```

```python
plt.figure(figsize=(15, 5))
ax = sns.barplot(x=top_9.index , y="ingresos insuficientes (%)", data=top_9)

plt.xticks(rotation=20)
ax.set(xlabel='Municipio')
plt.title('Hogares con insuficiencia de ingresos en los municipios del Cauca')
plt.savefig("socioeconomicos-Cauca.png")
plt.show()
```

```python
socioeconomicos_df.describe()[['ingresos insuficientes (%)', 'hogares con energia (%)', 'hogares con acueducto (%)']].T.to_excel('Estadisticas_socioeconomicas_Cauca2.xlsx')
```
