```python
def slope(x1, y1, x2, y2):
    
    m = (y2-y1)/(x2-x1)
    
    b = y1 - m*x1
    
    return m, b
```

# Model Normalized Values

### select features and labels

```python
# Uncomment to see models with original data

"""
gfsi = pd.read_excel(Path + 'FoodSecurityIndex/Data_Countries/2020_GFSI_Data_Table_Data.xlsx', skiprows=5)
gfsi.drop(columns = ['Unnamed: 0', 'Unnamed: 2'], inplace = True)
gfsi.rename(columns = {'Unnamed: 1' : 'features'},inplace = True)
gfsi = gfsi[gfsi['Average (mean)']  != '.']
gfsi.index = gfsi.features
gfsi.drop(columns = ['Average (mean)', 'Average (median)', 'Minimum', 'Maximum', 'features'], inplace = True)
gfsi = gfsi.T
"""

# Take y (index)
y = scores_country['OVERALL FOOD SECURITY ENVIRONMENT']
# Take x
X = gfsi
```

### Train test split

```python
### Train-Test Split Normalized Values
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
X_train.shape , X_test.shape
```

### Create error function (MAE)

```python
from sklearn.metrics import mean_absolute_error
#from sklearn.metrics import mean_squared_error

def calculate_mae(model):
    # Predict in test data
    y_pred = model.predict(X_test)

    print(f'The mean absolute error is: {mean_absolute_error(y_test, y_pred)}')
    
```

### Linear Regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression

linear_reg_norm = LinearRegression().fit(X_train, y_train)


calculate_mae(linear_reg_norm)
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge

ridge_reg_norm = Ridge().fit(X_train, y_train)

calculate_mae(ridge_reg_norm)
```

### Lasso

```python
from sklearn.linear_model import Lasso

lasso_reg_norm = Lasso(alpha=0.1).fit(X_train, y_train)

calculate_mae(lasso_reg_norm)
```

### Elastic Net

```python
from sklearn.linear_model import ElasticNet

elastic_net_reg_norm = ElasticNet(alpha=0.1, l1_ratio=0.7).fit(X_train, y_train)

calculate_mae(elastic_net_reg_norm)
```

### MLP

```python
from sklearn.neural_network import MLPRegressor

# model
model = MLPRegressor(hidden_layer_sizes=(30, 20, 10, 5),random_state=1, max_iter=1000, batch_size=64).fit(X_train, y_train) 

calculate_mae(model)
```

### SVM

```python
from sklearn.svm import SVR

# Fit regression model
svr_poly = SVR(kernel='poly', gamma='auto', degree=1, epsilon=.1, coef0=1).fit(X_train, y_train)

calculate_mae(svr_poly)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rfmodel = RandomForestRegressor(max_depth=150, random_state=1).fit(X_train, y_train)

calculate_mae(rfmodel)
```

```python
def convertir_sequia(x):
    if x < 0.2:
        x = 0
    elif x >= 0.2 and x < 0.4:
        x = 1
    elif x >= 0.4 and x < 0.6:
        x = 2
    elif x >= 0.6 and x < 0.8:
        x = 3
    elif x >= 0.8:
        x = 4
    return x
```

```python
def normalization_desastres_naturales(x):
    if x <= 1.83:
        x = 0
    elif x >= 17.78:
        x = 100
    else:
        x = slope(1.83, 0, 17.78, 100)[0] * x + slope(1.83, 0, 17.78, 100)[1]
    return x
```

```python
# '4.5) Sensitivity' : No -> GFSI

#'4.5.1) Food import dependency'
# https://www.cccauca.org.co/sites/default/files/archivos/perfil_del_comercio_exterior_del_cauca_2011_2012.pdf
columns.append('food_import_dependency')
dimension4.append('food_import_dependency')

gfsi_cauca['food_import_dependency'] = 60
gfsi_cauca_normalized['food_import_dependency'] = gfsi_cauca['food_import_dependency'].apply(normalization4_5_1)


#'4.5.2) Dependence on natural capital'
# https://www.dane.gov.co/index.php/estadisticas-por-tema/cuentas-nacionales/cuentas-nacionales-departamentales
# https://www.dane.gov.co/files/icer/2014/ICER_Cauca2014.pdf

columns.append('Dependence_natural_capital')
dimension4.append('Dependence_natural_capital')

# natural_resources (agrictura + aceites) / total
gdp_cauca = ((1565 + 351)/(2578 + 2569 + 1565 + 1677 + 1169 + 1214 + 1146 + 635 + 400 + 239 + 188 + 165 + 351)) * 100

gfsi_cauca['Dependence_natural_capital'] = gdp_cauca
gfsi_cauca_normalized['Dependence_natural_capital'] = gfsi_cauca['Dependence_natural_capital'].apply(normalization4_5_2)


gfsi_cauca
```

```python
# '4.7) Demographic stress' : ProyecciÃ³n de crecimiento de poblaciÃ³n en 5 aÃ±os

#'4.7.1) Projected population growth',
columns.append('Projected_population_growth')
dimension4.append('Projected_population_growth')

population_growth = ((food_security_df['Population 2019'] - food_security_df['Population 2015'])/food_security_df['Population 2015'])*100

gfsi_cauca['Projected_population_growth'] = population_growth
gfsi_cauca_normalized['Projected_population_growth'] = gfsi_cauca['Projected_population_growth'].apply(normalization4_7_1)

#'4.7.2) Urban absorption capacity'
# https://www.dane.gov.co/index.php/estadisticas-por-tema/cuentas-nacionales/cuentas-nacionales-departamentales
columns.append('Urban_absorption_capacity')
dimension4.append('Urban_absorption_capacity')

# Buscar PIB por municipio
# PIB 2014 - PIB 2019
crecimiento_PIB = ((15.623 - 13.899)/  15.623)* 100

urban_absorption = crecimiento_PIB - population_growth

gfsi_cauca['Urban_absorption_capacity'] = urban_absorption
gfsi_cauca_normalized['Urban_absorption_capacity'] = gfsi_cauca['Urban_absorption_capacity'].apply(normalization4_7_2)

gfsi_cauca
```

## Dimensional Analisis

```python
len_dim1 = len(dimension1)
len_dim2 = len(dimension1) + len(dimension2)
len_dim3 = len(dimension3) + len_dim2
len_dim4 = len(dimension4)
def get_dimnesion_score(dimension, model, df, dim):
    if dim == 1:
        weights = model.coef_[0:len(dimension)]
    elif dim == 2:
        weights = model.coef_[len_dim1 : len_dim1 + len(dimension)]
    elif dim == 3:
        weights = model.coef_[len_dim2 : len_dim2 + len(dimension)]
    elif dim == 4:
        weights = model.coef_[-len(dimension) : ]
    
    #print(weights)
    dimension_df = df[dimension]
    #print(dimension_df)
    # multiply
    dimension_weigth = dimension_df * weights
    #print(dimension_weigth)
    # calculate mean
    return dimension_weigth.sum(axis=1)
```

```python
dimension1_coef = get_dimnesion_score(dimension1, linear_reg_norm, gfsi_cauca_normalized, 1)
dimension1_coef
```
