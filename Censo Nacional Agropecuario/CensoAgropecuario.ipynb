# Pre-Processing

### Credit data:

```python
# Drop Na
CensoAgropecuario = CensoAgropecuario[CensoAgropecuario['P_S11P136'].notna()]
CensoAgropecuario
```

```python
CensoAgropecuario.isna().sum()
```

```python
print(f'The values of credit are: {set(CensoAgropecuario.P_S11P136.to_list())}')
# Credit? yes = 1 no = 2 -> yes = 1 , no = 0
CensoAgropecuario.P_S11P136 = CensoAgropecuario.P_S11P136.replace(2.0, 0.0)
print(f'The values of credit are:  {set(CensoAgropecuario.P_S11P136.to_list())}')
CensoAgropecuario = CensoAgropecuario.rename(columns={'P_S11P136':'Credito_Solicitado'})
```

```python
print(f'The values of credit are: {set(CensoAgropecuario.P_S11P136A.to_list())}')
# approved? yes = 1, no = 2, no credit = 0 -> yes = 1 , no, no credit = 0
CensoAgropecuario.P_S11P136A = CensoAgropecuario.P_S11P136A.replace(2.0, 0.0)
print(f'The values of credit are:  {set(CensoAgropecuario.P_S11P136A.to_list())}')
CensoAgropecuario = CensoAgropecuario.rename(columns={'P_S11P136A':'Credito_Aprobado'})
```

```python
credit = ['P_S11P136B_SP1', 'P_S11P136B_SP2', 'P_S11P136B_SP3', 'P_S11P136B_SP4', 'P_S11P136B_SP5', 'P_S11P136B_SP6', 'P_S11P136B_SP7', 'P_S11P136B_SP8']
CensoAgropecuario[credit] = CensoAgropecuario[credit].fillna(0)
CensoAgropecuario['Credito_banco'] = CensoAgropecuario['P_S11P136B_SP1'] + CensoAgropecuario['P_S11P136B_SP2'] + CensoAgropecuario['P_S11P136B_SP3'] + CensoAgropecuario['P_S11P136B_SP4']
CensoAgropecuario['Credito_ong'] = CensoAgropecuario['P_S11P136B_SP5'] + CensoAgropecuario['P_S11P136B_SP8']
CensoAgropecuario['Credito_gobierno'] = CensoAgropecuario['P_S11P136B_SP6']
CensoAgropecuario['Credito_onu'] = CensoAgropecuario['P_S11P136B_SP7']
CensoAgropecuario.drop(columns=credit, inplace=True)
CensoAgropecuario
```

### Investments Data

```python
inversion = ['P_S11P137_SP1', 'P_S11P137_SP2', 'P_S11P137_SP3', 'P_S11P137_SP4', 'P_S11P137_SP5', 'P_S11P137_SP6', 'P_S11P137_SP7', 'P_S11P137_SP8', 'P_S11P137_SP10', 'P_S11P137_SP11']
CensoAgropecuario[inversion] = CensoAgropecuario[inversion].fillna(0)
CensoAgropecuario['inversion_tecnologia'] = CensoAgropecuario['P_S11P137_SP3'] + CensoAgropecuario['P_S11P137_SP4'] + CensoAgropecuario['P_S11P137_SP10']
CensoAgropecuario['inversion_otros'] = CensoAgropecuario['P_S11P137_SP1'] + CensoAgropecuario['P_S11P137_SP2'] + CensoAgropecuario['P_S11P137_SP5'] + CensoAgropecuario['P_S11P137_SP6'] + CensoAgropecuario['P_S11P137_SP7'] + CensoAgropecuario['P_S11P137_SP8'] + CensoAgropecuario['P_S11P137_SP11'] 
CensoAgropecuario.drop(columns=inversion, inplace=True)
CensoAgropecuario
```

### Area usage data

```python
areas = ['P_S12P142', 'P_S12P143', 'P_S12P144', 'P_S12P145', 'P_S12P146','P_S12P147', 'P_S10P121']
CensoAgropecuario['area_cultivo'] = CensoAgropecuario['P_S12P142'] 
CensoAgropecuario['area_bosques'] = CensoAgropecuario['P_S12P146'] 
CensoAgropecuario['area_nocultivada'] = CensoAgropecuario['P_S12P143'] + CensoAgropecuario['P_S12P144'] + CensoAgropecuario['P_S12P145'] 
CensoAgropecuario['area_infraestructura'] = CensoAgropecuario['P_S12P147']

print(f'The values of areas are:  {set(CensoAgropecuario.P_S10P121.to_list())}')
CensoAgropecuario['P_S10P121'] = CensoAgropecuario['P_S10P121'].replace(9.0, 0.0)
CensoAgropecuario['P_S10P121'] = CensoAgropecuario['P_S10P121'].replace(2.0, 0.0)
print(f'The values of areas are:  {set(CensoAgropecuario.P_S10P121.to_list())}')
CensoAgropecuario['infraestructura_en_construccion'] = CensoAgropecuario['P_S10P121']

CensoAgropecuario.drop(columns=areas, inplace=True)
CensoAgropecuario
```

### Water Sources Data

```python
fuentes_agua = ['P_S11P124_SP1', 'P_S11P124_SP2', 'P_S11P124_SP3', 'P_S11P124_SP4', 'P_S11P124_SP5','P_S11P124_SP6', 'P_S11P124_SP7', 'P_S11P124_SP8', 'P_S11P124_SP9','P_S11P124_SP10', 'P_S11P124_SP11']
CensoAgropecuario[fuentes_agua] = CensoAgropecuario[fuentes_agua].fillna(0)

CensoAgropecuario['agua_lluvia'] = CensoAgropecuario['P_S11P124_SP1'] 
CensoAgropecuario['agua_fuente_natural'] = CensoAgropecuario['P_S11P124_SP2'] + CensoAgropecuario['P_S11P124_SP3'] + CensoAgropecuario['P_S11P124_SP4'] + CensoAgropecuario['P_S11P124_SP5'] + CensoAgropecuario['P_S11P124_SP6'] + CensoAgropecuario['P_S11P124_SP7'] 

for i in range (2,6):
      CensoAgropecuario['agua_fuente_natural'] = CensoAgropecuario['agua_fuente_natural'].replace(i, 1.0)

CensoAgropecuario['agua_no_acceso'] = CensoAgropecuario['P_S11P124_SP11']
CensoAgropecuario['agua_acueducto'] = CensoAgropecuario['P_S11P124_SP8']
CensoAgropecuario['agua_carrotanque'] = CensoAgropecuario['P_S11P124_SP9']
CensoAgropecuario['distrito_riego'] = CensoAgropecuario['P_S11P124_SP10']


CensoAgropecuario.drop(columns=fuentes_agua, inplace=True)
CensoAgropecuario
```

```python
dificultades_agua = ['P_S11P126_SP4', 'P_S11P126_SP7', 'P_S11P126_SP3', 'P_S11P126_SP5', 'P_S11P126_SP1', 'P_S11P126_SP2']

CensoAgropecuario[dificultades_agua] = CensoAgropecuario[dificultades_agua].fillna(0)

CensoAgropecuario['no_agua_contaminacion'] = CensoAgropecuario['P_S11P126_SP1']

CensoAgropecuario['no_agua_desastres_naturales'] = CensoAgropecuario['P_S11P126_SP7'] + CensoAgropecuario['P_S11P126_SP2']
CensoAgropecuario['no_agua_desastres_naturales'] = CensoAgropecuario['no_agua_desastres_naturales'].replace(2.0, 1.0)

CensoAgropecuario['no_agua_sequia'] = CensoAgropecuario['P_S11P126_SP4']

CensoAgropecuario['no_agua_fallos_servicio'] = CensoAgropecuario['P_S11P126_SP3'] + CensoAgropecuario['P_S11P126_SP5']
CensoAgropecuario['no_agua_fallos_servicio'] = CensoAgropecuario['no_agua_desastres_naturales'].replace(2.0, 1.0)

CensoAgropecuario.drop(columns=dificultades_agua, inplace=True)
CensoAgropecuario
```
